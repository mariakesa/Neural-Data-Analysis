```python
import numpy as np
from scipy import io
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dat=io.loadmat('/home/maria/Documents/NeuralDataAnalysis/Neural-Data-Analysis/MITCompNeuroCourse/fish.mat')
print(dat)
```

```python
freqs=np.fft.fft(dat['stim'].flatten())
print(freqs)
```

```python
npnts=dat['stim'].flatten().shape[0]
signal_amp=2*np.abs(freqs)/npnts
print(signal_amp)
#Get rid of DC component
plt.plot(signal_amp[1:])
plt.show()
print(dat['time'])
srate=0.5/1000
hz=np.linspace(0,srate/2,np.floor(npnts/2)+1)
#Get rid of DC component
plt.plot(hz[1:1000],signal_amp[0:hz.shape[0]][1:1000])
```
