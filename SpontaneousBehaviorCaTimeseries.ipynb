```python
import numpy as np
from scipy import io
import matplotlib.pyplot as plt
```

```python
data=io.loadmat('/home/maria/Documents/NeuralDataAnalysis/spont_M161025_MP030_20161120.mat')['Fsp']
```

```python
#Data is neurons x timepoints
print(data.shape)
```

```python
plt.plot(data[0,:100])
```

```python
#Running mean filter
def running_mean(X_neuron,k):
    print(X_neuron.shape)
    filtered_neuron=[]
    for i in range(k+1,X_neuron.shape[0]-k-1):
        filtered_neuron.append(np.mean(X_neuron[i-k:i+k]))
    print(len(filtered_neuron))
    return filtered_neuron

rm_3=running_mean(data[0,:],3)[0:100]
rm_10=running_mean(data[0,:],10)[0:100]
rm_100=running_mean(data[0,:],100)[0:100]
plt.plot(rm_3)
plt.plot(rm_10)
plt.plot(rm_100)
```
